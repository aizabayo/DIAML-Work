```python
print("Given data")
Investment = 100000              # Initial investment
Customers = 100                  # Number of customers per day
grow_rate = 1/100                # Daily customer growth rate
profit = 10                      # Daily profit provided by each customers
print(f"Initial investment  = {Investment} $")
print(f"Number of customer = {Customers}")
print(f"Grow rate of customer per day = {grow_rate}")
print(f"Profit provided by each customers = {profit} $")
#Initializing the variable
number_day = 0                   
Total_profit = 0                # Profit before start
days = []                        # List Variable to store number of days
cumulated_profit = []            # Variable to store profit of daily to be used

while Total_profit < Investment:
    Customer_increase = Customers * grow_rate      # Customer grow per day
    Customers = Customers + Customer_increase      # Increment day by day based on grow rate
    day_profit = Customers * profit                # Profit generated per day
    Total_profit = Total_profit + day_profit       # Incrementing the profit with day profit
    number_day = number_day + 1
    days.append(number_day)
    cumulated_profit.append(Total_profit)
    
print(f"The total number of days required to accumulate the initial investment is {number_day} days")
           # Variable to Store profit Earned daily

#Interpolation to get day when profit was $100,000
check  = interpolate.interp1d(cumulated_profit, days, kind = 'linear') # function used to check the exact number of days when 100,000$ profit found
exact_day = int(check(100000))   # variable that convert value into integer to help in estimation of number of days.                             #
#Ploting codes the figure where days is on x_axis and cumulated profit on y -axis
plt.plot(days, cumulated_profit, 'b' , marker = '.', label = 'profit Iteration')
plt.axhline(y = Investment ,color ='g', linestyle = '--', label = 'Initial Investment')   # creating horizontal green  dashed line on y-axis
plt.axvline(x = number_day, linestyle = '--', color = 'r', label = 'Breakeven day')       # creating vertical red dashed line on x-axis
#codes to create function that will add legend of approximate date inital investment will occur.
plt.annotate(f"Initial Investment ({Investment} $)", xy = (number_day + 5, Investment + 1000))
plt.annotate(f"Breakeven day ({number_day} days)", xy = (number_day + 5, Investment - 8000))
#creating a point to indicate the time the initial investment is covered based on interpolation
plt.scatter(exact_day, 100000,  color = 'k', marker='o', label=f'Profit = 100,000 $(Day {exact_day})')
plt.title("Cumulated profit per day curve")
plt.xlabel("Days")
plt.ylabel("Cumulated Profit")
plt.legend(loc = 4)

plt.show()
print(f"The day when profit  correspond to 100,000 $ is day {exact_day}")
```

```python
df = pd.read_excel('ebola_download.xlsx') # loading the datafile using pandas
print("Given Data\n")
print(df)
```

```python
print("Exctracted Data\n")
print(df[['Date','Cases', 'Death']])    # Extracting Date,Case, and Death from dataset 
print("The Generete Missing date between first and last date\n")
First_date = df['Date'].min()
End_date = df['Date'].max()
Missing_Date= pd.date_range(start = First_date, end = End_date).difference(df['Date'])    # declare Variable  of  missing date and insterted range of starting and ending date of dataset usie difference to remove the existing date in Date
data = {'Date': Missing_Date }                # Declkaring variable 'data' to store the missing date
Data = pd.DataFrame(data)                     # Creating dataframe of missing date
print("Missing date")
Data      # calling data frame of missing date
```

```python
# Merging and Sorting two dataFrame and create one dataset  and then sort from starting to the End date
Merge = pd.concat([df,Data]).sort_values(by = 'Date').reset_index(drop= True)         #Merging  two dataset 
print('the Original DataFrame is Merged with Missing Date Dataframe\n\n',Merge)
print('updated data file\n')
Merge.to_excel('new_ebola_download.xlsx') 
print("Filling the missing values linearly of 'Cases','Death'\n")
#Use linear interpolation to get missing values of Cases and Deaths linearly
Merge['Cases'].interpolate(method = 'linear', inplace = True)   
Merge['Death'].interpolate(method = 'linear', inplace= True)
print(Merge)
```

```python
#Extraction of indices where value are grater than certain range
Limits = [100,500,1000,2000,5000]   # Declare  to be exceeded 
Cases_indices = {}  # create dictionary for storing cases indices
Death_indices = {}   # create dictionary for storing Death indices
print('\n\t\t Exceedance Cases\n')
for Limit in Limits:
    indices = Merge.index[Merge['Cases'] >Limit]
    if len(indices)> 0:
        Cases_indices[Limit] = indices[0]
# Extract and print the exact number of Cases for each threshold
for Limit, index in Cases_indices.items():
    exact_cases = Merge.loc[index, 'Cases'] 
    print(f"For Cases > {Limit},the first exceedance is on {index} index, the corresponding date is {Merge.loc[index, 'Date']} with {exact_cases} cases")
        
# Ploting the merged data
plt.plot(Merge['Date'], Merge['Cases'], 'b', label = 'Cases (Interpolated)') 
plt.plot(Merge['Date'], Merge['Death'], 'g', linestyle = '-' ,label = 'Death (Interpolated)') 
        
for Limit, index in Cases_indices.items():
    plt.scatter(Merge.loc[index, 'Date'], Merge.loc[index, 'Cases'], c= 'r', marker = 'o',label = f'Cases>{Limit}')

for Limit in Limits:
    indices = Merge.index[Merge['Death'] > Limit]
    if len(indices)> 0:
        Death_indices[Limit] = indices[0]
# Extract and print the exact number of Death for each threshold
print("\n\t\t Exceedance Deaths \n")
for Limit, index in Death_indices.items():
    exact_Death = Merge.loc[index, 'Death'] 
    print(f"For Deaths > {Limit},the first exceedance is on {index} index, the corresponding date is {Merge.loc[index, 'Date']} with {exact_Death} numbers")
        
# Ploting the merged data
plt.plot(Merge['Date'], Merge['Cases'], 'b', label = 'Cases (Interpolated)') 
plt.plot(Merge['Date'], Merge['Death'], 'g', linestyle = '-', label = 'Death (Interpolated)') 
        
for Limit, index in Death_indices.items():
    plt.scatter(Merge.loc[index, 'Date'], Merge.loc[index, 'Death'], c= 'k', marker = 'o',label = f'Cases>{Limit}')
plt.xlabel('Date')
plt.ylabel("Ebola Cases and Deaths")
plt.title("Ebola Cases-Death curve against Date")
plt.legend()
plt.show()
```
